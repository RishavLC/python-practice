```python
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 5]
y = [5, 33, 50]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# marker
x = [1, 3, 5]
y = [5, 33, 50]

plt.plot(x, y, marker='X', markersize='14',mec='r',mfc='g')
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# limestyle
# marker
x = [1, 3, 5]
y = [5, 33, 50]

plt.plot(x, y, marker='X', markersize='14',mec='r',mfc='g',linestyle=':', color='m')
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# markerlinestyle color
x = [1, 3, 5]
y = [5, 33, 50]

plt.plot(x, y,'o--y')
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# subplot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot([1, 2, 3], [4, 5, 6])

plt.subplot(1, 2, 2)  # second plot
plt.plot([1, 2, 3], [6, 5, 4])

plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Graph with Grid")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True, color='m', linestyle='--', linewidth=1)
plt.show()
```

```python
# scatterplot
x=[3,7,21,13,45]
y=[12,3,54,32,27]
plt.scatter(x,y,alpha=1,color='m',cmap='magma')
plt.colorbar()
```

```python
names = ['Ram', 'Shyam', 'Gita']
marks = [85, 90, 95]

plt.bar(names, marks)
plt.title("Marks of Students")
plt.xlabel("ioh")
plt.ylabel("gfg")
plt.show()
```

```python
import numpy as np

data = np.random.normal(65, 15, 100)

plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sizes = [30, 20, 25, 25]
labels = ['Python', 'Java', 'C++', 'HTML']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Language Usage")
plt.show()
```

```python
data = [80, 85, 90, 75, 95, 100, 70]
plt.boxplot(data)
plt.title("Box Plot of Marks")
plt.show()
```
